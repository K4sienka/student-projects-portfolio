```python
print('Liczba wszystkich pasażerów: {}'.format(len(df)))
```

Zmienne 'sex', 'embarked', 'pclass', 'survived', 'sibsp' i 'parch' mają niską kardynalność, więc łatwo je analizować. 'cabin', 'boat' i 'body' mają dużo unikalnych wartości, co może utrudniać pracę. Takie zmienne warto uprościć lub pominąć. 'body' i 'boat' raczej odrzuciłabym, bo ciężko je sensownie zakategoryzować. 'sibsp' i 'parch' można pogrupować na kategorie według liczby osób (ale też bym raczej pominęła).

```python
print('Ilość kabin:',len(np.array(df['cabin'].unique())))
```

```python
df['CabinReduced'] = df['cabin'].astype(str).str[0]
df[['cabin', 'CabinReduced']].head(20)
```

```python
print('Etykiety dla CabinReduced: {}'.format(len(df['CabinReduced'].unique())))
print('Etykiety dla cabin: {}'.format(len(df['cabin'].unique())))
print('\n')

precent = ((len(df['cabin'].unique()) - len(df['CabinReduced'].unique())) / len(df['cabin'].unique()))*100
print(f'Zredukowano o {precent:.2f}%') 
```

Ilość kabin, która była zajmowana w trakcie podróży jest bardzo wysoka. Utrudnia to dalsze analizy, dlatego w celu ułatwienia pracy, należy w jakiś sposób zredukować kardynalność tej zmiennej. Uważam, że przyjęcie, aby zredukować nazwy kabin tylko do pierwszej litery jest dużym uproszczeniem. Wydaje mi się, że kolejne litery alfabetu oznaczają "piętro", na którym dane kabiny się znajdowały, ewentualnie może jakieś sektory. Taka informacja jest dla nas jakąś wskazówką o ich położeniu i zależności między sobą. Dzięki tej redukcji z 187 etykiet przeszliśmy tylko na 9 co ułatwia dalszą pracę z takimi danymi. Jedyny negatywny skutek jaki mi przychodzi do głowy przy takich zmianach, to utrata bardziej szczegółowych informacji o lokalizacji pasażerów, np: ile osób znajdowało się w konkretnej kabinie.

```python
df = pd.read_csv("titanic.txt", header=None, na_values='?')
df.columns = [
    'pclass', 'survived', 'name', 'sex', 'age', 'sibsp', 'parch', 'ticket',
    'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest'
]
df['CabinReduced'] = df['cabin'].where(df['cabin'].notna()).astype(str).str[0]
df.to_csv("titanic2.csv", index=False)
```
