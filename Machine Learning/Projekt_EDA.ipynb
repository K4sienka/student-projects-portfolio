```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
file_path = "rejestr_sor30.06.2022.xlsx"
df = pd.read_excel(file_path)

# Podstawowe informacje o danych
print("Rozmiar danych (wiersze, kolumny):", df.shape)
display(df.head(5))
```

```python
new_column_names = {
    'Nr zezwolenia na dopuszczenie do obrotu środka ochrony roślin': 'nr_zezwolenia',
    'Nazwa środka ochrony roślin': 'nazwa_srodka',
    'Rodzaj środka': 'rodzaj_srodka',
    'Nazwa, siedziba i adres podmiotu, który uzyskał zezwolenie': 'podmiot',
    'Producent środka': 'producent',
    'Zawartość, nazwa zwyczajowa substancji czynnej środka ochrony roślin': 'substancje_czynne',
    'Klasyfikacja środka ochrony roślin pod względem stwarzania przez niego zagrożeń dla zdrowia człowieka lub pełna klasyfikacja środka zgodna z rozp. 1272/2008': 'zagrozenie_zdrowie',
    'Klasyfikacja środka ochrony roślin pod względem stwarzania przez niego zagrożeń dla pszczół': 'zagrozenie_pszczoly',
    'Klasyfikacja środka ochrony roślin pod względem stwarzania przez niego zagrożeń dla organizmów wodnych': 'zagrozenie_wodne',
    'Klasyfikacja środka ochrony roślin pod względem stwarzania przez niego zagrożeń dla środowiska': 'zagrozenie_srodowisko',
    'Termin ważności zezwolenia': 'termin_waznosci',
    'Okres na zużycie istniejących zapasów środka ochrony roślin dla sprzedaży i dystrybucji': 'okres_sprzedaz',
    'Okres na zużycie istniejących zapasów środka ochrony roślin dla unieszkodliwiania, przechowywania i stosowania': 'okres_unieszkodliwianie',
    'Rodzaje opakowań (rodzaj opakowania, pojemność, materiał)': 'rodzaj_opakowania',
    'Uwagi': 'uwagi'
}

df = df.rename(columns=new_column_names)
df.head(5)
```

```python
missing_values = df.isna().sum().reset_index()
missing_values.columns = ['Kolumna', 'Liczba braków']
missing_values['Procent braków'] = (missing_values['Liczba braków'] / len(df)) * 100
missing_values = missing_values.sort_values(by='Liczba braków', ascending=False)
display(missing_values)
```

```python
unique_counts = df.nunique().reset_index()
unique_counts.columns = ['Kolumna', 'Liczba unikalnych wartości']
unique_counts = unique_counts.sort_values(by='Liczba unikalnych wartości', ascending=False)
display(unique_counts)
```

```python
columns_to_check = ['rodzaj_srodka', 'producent', 'podmiot', 'zagrozenie_zdrowie', 'substancje_czynne']

for col in columns_to_check:
    print(f"\nNajczęstsze wartości w kolumnie: {col}")
    display(df[col].value_counts().head(5))
```

```python
print("\nNajczęstsze podmioty uzyskujące zezwolenie:")
podmiot_counts = df['podmiot'].value_counts()
display(podmiot_counts[podmiot_counts>10])
print('\n')
```

```python
# Jeśli są NaN, zamieniamy na pusty tekst
df['zagrozenie_zdrowie'] = df['zagrozenie_zdrowie'].fillna('')

# Funkcja kategoryzująca
def check_h_category(codes, category_prefix):
    return int(any(code.startswith(category_prefix) for code in codes))

# Rozdzielamy kody na listę
df['zagrozenie_zdrowie_split'] = df['zagrozenie_zdrowie'].apply(lambda x: [code.strip() for code in x.split(',') if code.strip()])

# Tworzymy trzy nowe kolumny:
df['zagrozenie_zdrowotne'] = df['zagrozenie_zdrowie_split'].apply(lambda x: check_h_category(x, 'H3'))
df['zagrozenie_srodowiskowe'] = df['zagrozenie_zdrowie_split'].apply(lambda x: check_h_category(x, 'H4'))
df['zagrozenie_fizyczne'] = df['zagrozenie_zdrowie_split'].apply(lambda x: check_h_category(x, 'H2'))

# Usuwamy pomocniczą kolumnę
df = df.drop(columns=['zagrozenie_zdrowie_split'])

# Podejrzenie efektu
display(df[['nazwa_srodka', 'zagrozenie_zdrowie', 'zagrozenie_zdrowotne', 'zagrozenie_srodowiskowe', 'zagrozenie_fizyczne']].head())
```

```python
df['zagrozenie_zdrowie'] = df['zagrozenie_zdrowie'].fillna('')

def map_h_categories(codes):
    categories = []
    if any(code.startswith('H3') for code in codes):
        categories.append('zdrowotne')
    if any(code.startswith('H4') for code in codes):
        categories.append('środowiskowe')
    if any(code.startswith('H2') for code in codes):
        categories.append('fizyczne')
    if categories:
        return ', '.join(categories)
    else:
        return 'brak'

df['zagrozenie_zdrowie_split'] = df['zagrozenie_zdrowie'].apply(lambda x: [code.strip() for code in x.split(',') if code.strip()])
df['kategoria_zagrozenia'] = df['zagrozenie_zdrowie_split'].apply(map_h_categories)
df = df.drop(columns=['zagrozenie_zdrowie_split'])
display(df[['nazwa_srodka', 'zagrozenie_zdrowie', 'kategoria_zagrozenia']].head(10))
```

```python
# 1. Liczenie liczby środków w każdej kategorii zagrożenia
kategorie_counts = df['kategoria_zagrozenia'].value_counts().reset_index()
kategorie_counts.columns = ['kategoria_zagrozenia', 'liczba_srodkow']

# Wyświetlenie tabelki
display(kategorie_counts)

# 2. Wykres: ile środków w każdej kategorii
plt.figure(figsize=(10, 6))
sns.barplot(data=kategorie_counts, x='liczba_srodkow', y='kategoria_zagrozenia', hue='kategoria_zagrozenia', palette='viridis')
plt.title('Liczba środków ochrony roślin w każdej kategorii zagrożenia')
plt.xlabel('Liczba środków')
plt.ylabel('Kategoria zagrożenia')
plt.grid(True, axis='x')
plt.show()
```

```python
# 1. Konwersja kolumn z datami na typ datetime
date_columns = ['termin_waznosci', 'okres_sprzedaz', 'okres_unieszkodliwianie']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')  # zamienia na datę, jak coś nie czytelne to NaT

# 2. Dodanie nowej kolumny: dni do końca ważności
import datetime

# Bierzemy dzisiejszą datę
today = datetime.datetime.today()

# Liczymy ile dni zostało
df['dni_do_waznosci'] = (df['termin_waznosci'] - today).dt.days

# Podgląd efektu
display(df[['nazwa_srodka', 'termin_waznosci', 'dni_do_waznosci']].head(10))
```

```python
# mediana liczby dni do wygaśnięcia
median_dni = df['dni_do_waznosci'].mean()
print(f"Mediana dni do wygaśnięcia: {median_dni:.0f} dni")

# odsetek środków już po terminie
expired_pct = (df['dni_do_waznosci'] < 0).mean() * 100
print(f"Środki po terminie: {expired_pct:.1f}% wszystkich wpisów")
```

```python
import re

def extract_litres(text):
    match = re.search(r'(\d+)\s*[lL]', str(text))
    if match:
        return int(match.group(1))
    else:
        return None
#ile litrow srodka
df['pojemnosc_litry'] = df['rodzaj_opakowania'].apply(extract_litres)

# 2. Zgrupowanie po klasie ryzyka i policzenie statystyk pojemności
mean_litres = (
    df.groupby('kategoria_zagrozenia')['pojemnosc_litry']
      .agg(['count', 'mean', 'median'])
      .reset_index()
      .sort_values('mean')
)
print(mean_litres)
```

```python
# 3. Proste porównanie – barplot średnich
plt.figure(figsize=(8,5))
sns.barplot(data=mean_litres,
            x='mean',
            y='kategoria_zagrozenia',
            palette='viridis')
plt.xlabel('Średnia pojemność opakowania [l]')
plt.ylabel('Kategoria zagrożenia')
plt.title('Średnia pojemność vs. rodzaj zagrożenia')
plt.grid(axis='x')
plt.show()
```

```python
import pandas as pd

# 1. Udział rynkowy producentów (%)
producer_counts = df['producent'].value_counts()
producer_share = (producer_counts / len(df) * 100).reset_index()
producer_share.columns = ['producent', 'percent_market']
producer_share['percent_market'] = producer_share['percent_market'].round(2)

print("Udział rynkowy producentów (top 10):")
display(producer_share.head(10))
```

```python
# 2. Grupowanie producentów i sumowanie liczby groźnych preparatów
dangerous_by_producer = (
    df.groupby('producent')['zagrozenie_zdrowotne']
      .sum()
      .reset_index()
      .sort_values(by='zagrozenie_zdrowotne', ascending=False)
)

# 3. Wyświetlenie Top 10 producentów:
top10_dangerous = dangerous_by_producer.head(10)
print(top10_dangerous)
```

```python
# 1. Oblicz wartości
vc = df['kategoria_zagrozenia'].value_counts()

# 2. Połącz małe kategorie w 'Inne'
threshold = 0.05  # progiem np. 5%
major = vc[vc / vc.sum() >= threshold]
minor_sum = vc[vc / vc.sum() < threshold].sum()
major['Inne'] = minor_sum

# 3. Przygotuj explode tak, by „wybuchnąć” największy wycinek
explode = [0.1 if i==major.idxmax() else 0 for i in major.index]

# 4. Rysuj donut chart
fig, ax = plt.subplots(figsize=(7,7))
wedges, texts, autotexts = ax.pie(
    major.values,
    labels=major.index,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
    pctdistance=0.75
)
# dodaj środkowe "dziurę"
centre_circle = plt.Circle((0,0), 0.50, fc='white')
fig.gca().add_artist(centre_circle)

# Legenda na boku
ax.set_title('Procentowy udział kategorii zagrożeń')
plt.tight_layout()
plt.show()
```
