### Zadnie 2: Prezentacja danych w formie tabelarycznej [10 pkt]

Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

#### a) Całość danych
Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

 > Indeks: Year </br>
 > Wszystkie lata, dla których zostały wyznaczone wartości parametru GDP growth (annual %)
 
 > Kolumny: (-20, -15],	(-15, -10], ... , (30, 35] </br>
 > Kolejne przedziały wartości zdefiniowane jako `pandas.IntervalIndex`

```python
#usuwamy zbedne kolumny
data = recession_data_1.drop(columns=["Country Name", "Region Name", "Income Group"])

#tworzymy listę przedziałów wartości GDP growth jako IntervalIndex
bins = pd.IntervalIndex.from_tuples(
    [(-20, -15), (-15, -10), (-10, -5), (-5, 0), (0, 5),
     (5, 10), (10, 15), (15, 20), (20, 25), (25, 30), (30, 35)]
)

#pusty slownik
results = {}

#przechodzimy przez wszystkie kolumny z latami
for year in data.columns:
    #dane dla danego roku
    values = data[year]
    #grupuje do przedzialow (cut) i zlicza ile wartosci wpada w przedzial
    count = pd.cut(values, bins).value_counts()
    #sortujemy zliczenia zgodnie z kolejnością przedziałów
    sorted_count = count.sort_index()
    #dzielimy przez sume zeby miec proporcje
    proportions = sorted_count / sorted_count.sum()
    #zapisanie wyniku do słownika results
    results[year] = proportions

#tworzymy ramke danych z wynikow - lata to wiersze, przedzialy to kolumny
recession_table_1 = pd.DataFrame.from_dict(results, orient='index')
recession_table_1.index.name = "Year"

#dane mają byc z dwoma miejscami po przecinku
pd.options.display.float_format = "{:.2f}".format
display(recession_table_1)
```

```python
### Komórka testowa
# Kolumny
lower_boundaries = np.arange(-20, 35, 5)
for i in range(len(lower_boundaries)):
    assert recession_table_1.columns[i] == pd.Interval(lower_boundaries[i], lower_boundaries[i]+5, closed='right')
# Indeks
assert recession_table_1.index.name == "Year"
assert len(set(list(recession_table_1.index)) & set([2006, 2007, 2008, 2009, 2010, 2011])) == 6 \
    or len(set(list(recession_table_1.index)) & set(['2006', '2007', '2008', '2009', '2010', '2011'])) == 6
# Wartości w obrębie tablicy
assert recession_table_1.values.min() >= 0.0
assert recession_table_1.values.max() <= 1.0
assert np.all(np.isclose(np.array(recession_table_1.sum(axis=1).values, dtype=float), 1))
```

#### b) Podział na regiony geograficzne
Druga tablica (`recession_table_2`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego z rejonów geograficznych.

 > Indeks: 2-poziomowe nagłówki </br>

 >> Nagłówki 1 poziomu: Region Name </br>
 >> Nazwy regionów geograficznych

 >> Nagłówki 2 poziomu: Year </br>
 >> Wszystkie lata, dla których zostały wyznaczone wartości parametru GDP growth (annual %) powtórzone dla każdego z rejonów

 > Kolumny: (-20, -15],	(-15, -10], ... , (30, 35] </br>
 > Kolejne przedziały wartości zdefiniowane jako `pandas.IntervalIndex`

```python
#tworzymy listę przedziałów wartości GDP growth jako IntervalIndex
bins = pd.IntervalIndex.from_tuples(
    [(-20, -15), (-15, -10), (-10, -5), (-5, 0), (0, 5),
     (5, 10), (10, 15), (15, 20), (20, 25), (25, 30), (30, 35)]
)

#pusty slownik
results = {}

#regiony
regions = recession_data_1["Region Name"]

#przechodzimy przez wszystkie kolumny z latami
for region in regions.unique():
    #dane tylko dla danego regionu
    region_data = recession_data_1[regions == region]
    
    #niepotrzebne kolumny, zeby iterowac po latach
    region_data = region_data.drop(columns=["Country Name", "Region Name", "Income Group"])
    
    for year in region_data.columns:
        values = region_data[year]
        count = pd.cut(values, bins).value_counts()
        sorted_count = count.sort_index()
        proportions = sorted_count/sorted_count.sum()
        #klucz to krotki (region, year)
        results[(region, year)] = proportions

#tworzymy ramke danych z wynikow - lata to wiersze, przedzialy to kolumny
recession_table_2 = pd.DataFrame.from_dict(results, orient='index')
recession_table_2.index.names = ["Region Name", "Year"]

#dane mają byc z dwoma miejscami po przecinku
pd.options.display.float_format = "{:.2f}".format
display(recession_table_2)
```

```python
# Komórka testowa
# Kolumny
lower_boundaries = np.arange(-20, 35, 5)
for i in range(len(lower_boundaries)):
    assert recession_table_2.columns[i] == pd.Interval(lower_boundaries[i], lower_boundaries[i]+5, closed='right')
# Indeks
assert recession_table_2.index.names == ['Region Name', 'Year']
assert set(list(recession_table_2.index.levels[0])) == set(['Africa', 'Americas', 'Asia', 'Europe', 'Oceania'])
set(list(recession_table_2.index.levels[1]))
assert set(list(recession_table_2.index.levels[1])) == set([2006, 2007, 2008, 2009, 2010, 2011]) \
    or set(list(recession_table_2.index.levels[1])) == set(['2006', '2007', '2008', '2009', '2010', '2011'])
# # Wartości w obrębie tablicy
assert recession_table_2.values.min() >= 0.0
assert recession_table_2.values.max() <= 1.0
assert np.all(np.isclose(np.array(recession_table_2.sum(axis=1).values, dtype=float), 1))
```

### Zadanie 3: Prezentacja danych w formie graficznej [25 pkt]

#### a) Diagramy częstości

Bazując na wczytanych do notatnika danych przygotuj diagramy częstości odpowiadające danym zawartym w przygotowanych wcześniej tabelach (podział na identyczne przedziały, skonstruowane w oparciu o całość danych oraz indywidualnie dla każdego z rejonów geograficznych i dla każdego stopnia zamożności, osobne diagramy dla poszczególnych lat). 

Zestaw wyniki na pojedynczej figurze (siatka wykresów 2x5). W górnym wierszu zawrzyj diagram skonstruowany w oparciu o całość danych oraz cztery diagramy skonstruowane indywidualnie dla poszczególnych stopni zamożności, a w dolnym wierszu pięć dziagramów skonstruowanych indywidualnie dla poszczególnych rejonów geograficznych). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw identyczne zakresy wartości na odpowiadających sobie osiach wszystkich wykresów,

Uwaga! Zadanie należy wykonać w oparciu o wczytane do notatnika szeregi szczegółowe, a nie w oparciu o szeregi rozdzielcze częstości przygotowane w zadaniu 2.

```python
# Przygotowanie danych
years = [str(y) for y in range(2006, 2012)]
df_long = recession_data_1.melt(
    id_vars=["Country Name", "Region Name", "Income Group"],
    value_vars=years,
    var_name="Year",
    value_name="Value"
)

# Parametry histogramu
bins = np.arange(-20, 35, 5)
colors = {
    '2006': 'green',
    '2007': 'magenta',
    '2008': 'orange',
    '2009': 'red',
    '2010': 'blue',
    '2011': 'yellow'
}
ymax = 0.8

# Funkcja zliczająca częstość
def compute_frequencies(values):
    counts, _ = np.histogram(values, bins=bins)
    freq = counts / counts.sum() if counts.sum() > 0 else np.zeros_like(counts)
    return freq

# Tworzenie wykresu 2x5
fig, axes = plt.subplots(2, 5, figsize=(20, 10), sharex=True, sharey=True)

# Górny rząd: All + Income Groups
top_titles = ["All Data"] + list(df_long["Income Group"].unique())
for i, title in enumerate(top_titles):
    ax = axes[0, i]
    data = df_long if title == "All Data" else df_long[df_long["Income Group"] == title]
    for year in years:
        vals = data[data["Year"] == year]["Value"]
        freq = compute_frequencies(vals)
        ax.plot(bins[:-1], freq, label=year, color=colors[year])
    ax.set_title(title)
    ax.set_ylim(0, ymax)
    ax.set_xlim(bins[0], bins[-2])
    ax.set_xlabel("Value")
    ax.set_ylabel("Relative Frequency")
    ax.legend(title="Year", loc="upper right")
    ax.tick_params(axis='both', which='both', labelsize=10)
    ax.yaxis.grid(True)

# Dolny rząd: Regiony
bottom_titles = list(df_long["Region Name"].unique())[:5]
for i, title in enumerate(bottom_titles):
    ax = axes[1, i]
    data = df_long[df_long["Region Name"] == title]
    for year in years:
        vals = data[data["Year"] == year]["Value"]
        freq = compute_frequencies(vals)
        ax.plot(bins[:-1], freq, color=colors[year], label=year)
    ax.set_title(title)
    ax.set_ylim(0, ymax)
    ax.set_xlim(bins[0], bins[-2])
    ax.set_xlabel("Value")
    ax.set_ylabel("Relative Frequency")
    ax.legend(title="Year", loc="upper right")
    ax.tick_params(axis='both', which='both', labelsize=10)
    ax.yaxis.grid(True)
fig.suptitle("Porównanie częstości dla różnych danych", fontsize=16)
fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
## NASTEPNYM RAZEM UZYĆ HISTPLOT
```

#### c) Dyskusja wyników

Zastanów się jakie są mocne i słabe strony poszczególnych typów wykresów wykonanych w tym zadaniu. W jakich przypadkach wykorzystanie konkretnych typów wykresów do wizualizacji może sprawiać problemy?

Boxplot:
Zaletą boxplota jest to, że dobrze nadaje się do porównania rozkładu danych i wykrywania wartości odstających. Pokazuje medianę, kwartyle, jest dobry przy dużej ilości danych. Wadami jest to, że nie pokazuje pełnego rozkładu, może być mylący przy małych próbkach. W takiej sytuacji może ukrywać ciekawe właściwości danych, takie jak asymetria czy obecność kilku skupisk.

Violinplot:
Zaleta violinplota jest to, że łączy cechy boxplota i gęstości rozkładu, oprócz mediany i kwartylów pokazuje też kształt rozkładu np. czy dane są symetryczne, czy mają kilka "górek". Dobrze sprawdza się przy średnich albo dużych zbiorach danych. Wadą natomiast jest to, że przy małej liczbie obserwacji wykres jest niedokładnie przedstawiony. Może np: sztucznie wygładzać, przez co będzie wprowadzać w błąd. Przy dużej liczbie kategorii staje się on zatłoczony i mniej czytelny.

Swarmplot:
Zaletą swarmplota jest to, że pokazuje wszystkie dane punkt po punkcie. Jest przydatny przy małych i średnich zbiorach danych - ukazuje dokładnie rozrzut i zagęszczenie. Wadą jest to, że przy dużej liczbie punktów robi się nieczytelny przez nachodzenie się na siebie punktów lub może pominąć rysowanie niektórych punktów. Wpływa to zatem mocno na czytelność wizualizacji.

Stripplot:
Zaletą stripplota jest to, że jest prosty i szybki w wykonaniu, jest również dobrym zamiennikiem swarmplota przy dużej liczbie punktów. Dodatkowo łatwo można dodać kolory. Wadą  natomiast jest to, że punkty moga się nakładać, przez co tracimy informację o liczebności. Jest mniej czytelny niż swarmplot dla małych danych.
