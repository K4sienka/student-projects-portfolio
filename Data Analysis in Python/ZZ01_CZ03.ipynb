### Zadanie 4: Konstrukcja szeregów rodzielczych - bardziej skomplikowane tablice [10 pkt]

#### a) Forma tabelaryczna
Przygotuj szeregi rozdzielcze VEI w formie tablicy `vei_data` zawierającej następujące elementy:

> Indeks: VEI </br>
> Wartości VEI posortowane od najmniejszej do największej.

> Kolumny: 2-poziomowe nagłówki, dla każdego z nagłówków 1 poziomu powtarza się cały zestaw nagłówków 2 poziomu

>> Nagłówki 1 poziomu: All, Confirmed Eruption, Uncertain Eruption </br>
>> Całość danych oraz grupy wyznaczone na podstawie kolumny Eruption category</br>

>> Nagłówki 2 poziomu: f, cf, rf, crf </br>
>> Zliczenia poszczególnych kategorii (ang. frequency, absolute frequency), zliczenia skumulowane poszczególnych kategorii (ang. cumulative frequency, cumulative absolute frequency),  rf - częstość zliczeń poszczególnych kategorii (ang. relative frequency), crf - skumulowana częstość zliczeń poszczególnych kategorii (ang. cumulative relative frequency)

Wyświetl przygotowane tablicę w taki sposób, aby liczby w kolumnach f i cf oraz indeks tablicy VEI wyświetlały się bez miejsc po przecinku, a liczby z kolumn rf i crf z dokładnością do 4 miejsca po przecinku.

```python
# YOUR CODE HERE
pd.options.display.float_format = '{:.4f}'.format #wymuszenie 4 miejsca po przecinku przy wyświetlaniu

#funkcja z poprzedniego zestawu zadanie 3
def szeregi_rozdzielcze(df):
    vei_count = df['VEI'].value_counts().sort_index()  # Posortowane wartości VEI
    vei_count.index = vei_count.index.astype(int)  # VEI jako int
    
    f = vei_count.astype("Int64")  # Frequency, ma być Int64 dla obsługi NA
    cf = f.cumsum().astype("Int64")  # Cumulative frequency, Int64
    rf = f / f.sum()  # Relative frequency, 
    crf = rf.cumsum()  # Cumulative relative
   
    table = pd.DataFrame({
        'f': f,
        'cf': cf,
        'rf': rf.round(4),
        'crf': crf.round(4)     
    })

    return table


# Wszystkie erupcje
vei_all = szeregi_rozdzielcze(eruptions_data)

# Potwierdzone erupcje
confirmed = eruptions_data[eruptions_data['Eruption Category'] == 'Confirmed']
vei_confirmed = szeregi_rozdzielcze(confirmed)

# Niepotwierdzone erupcje (Uncertain)
unconfirmed = eruptions_data[eruptions_data['Eruption Category'] != 'Confirmed']
vei_uncertain = szeregi_rozdzielcze(unconfirmed)




# Tworzymy wielopoziomową tabelę
vei_data = pd.concat([vei_all, vei_confirmed, vei_uncertain],
                     axis=1,
                     keys=['All eruptions', 'Confirmed Eruptions', 'Uncertain Eruptions'])

vei_data
```

```python
## Komórka testowa
# Test 1: Indeks
assert (vei_data.index.name == 'VEI') == True
assert list(vei_data.index) == [0, 1, 2, 3, 4, 5, 6, 7]
# Test 2: Kolumny
assert len(set(list(vei_data.columns.levels[0])) & set(['All eruptions', 'Confirmed Eruptions', 'Uncertain Eruptions'])) == 3
assert set(list(vei_data.columns.levels[1])) == set(['f', 'cf', 'rf', 'crf'])
# Test 3: Wartości w kolumnach f
assert list(vei_data['All eruptions']['f'].values) == [1007, 1384, 3991, 1152, 511, 180, 53, 7]
assert list(vei_data['Confirmed Eruptions']['f'].values) == [803, 1227, 3727, 1133, 510, 180, 53, 7]
assert list(vei_data['Uncertain Eruptions']['f'].values[:5]) == [204, 157, 264, 19, 1] 
assert np.all(np.isnan(vei_data['Uncertain Eruptions']['f'].values[5:]))
```

#### b) Forma graficzna
Zwizualizuj dane z przygotowanych tablic za pomocą wykresów słupkowych zliczeń (lewa oś wykresu) z nałożonymi na nie łamanymi częstości skumulowanej (prawa oś wykresu). Zestaw wykresy na pojedynczej figurze (siatka wykresów 1x3). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo ustaw:
- identyczne zakresy wartości na odpowiadających sobie osiach wszystkich trzech wykresów,
- różne od domyślnych i od siebie nawzajem kolory wykresu słupkowego oraz łamanej częstości,
- kolory lewej i prawej osi oraz ich opisów identyczne z kolorem elementu wykresu, którego dotyczą,
- format wyświetlania wartości VEI bez miejsc po przecinku.

```python
def plot_vei_counts(vei_data_list, titles=("All", "Confirmed", "Unconfirmed")):
    fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

    #Różne kolory dla każdego wykresu słupkowego
    bar_colors = ['blue', 'green', 'orange']

    # Znajdowanie maksymalnego zakres osi X
    max_x_value = max(max(data.index) for data in vei_data_list)

    for ax, data, title, color in zip(axes, vei_data_list, titles, bar_colors):
        # Wykres słupkowy
        ax.bar(data.index, data['f'], color=color, alpha=0.7, label='Count')
        ax.set_ylabel("Liczność", color=color)
        ax.tick_params(axis='y', colors=color)

        # Ustawienie pełnego zakresu osi X
        ax.set_xticks(range(int(data.index.min()), int(max_x_value) + 1))
        ax.set_xticklabels(range(int(data.index.min()), int(max_x_value) + 1))

        ax.set_ylim(0, 4500)
       
         # Dodanie tytułu 
        fig.suptitle("Porównanie VEI - Liczność i Częstość Skumulowana", fontsize=16, fontweight='bold')
    
        # Wykres liniowy 
        ax2 = ax.twinx()
        ax2.plot(data.index, data['crf'],  linestyle='-', color='red', label='Cumulative Frequency')
        ax2.set_ylabel("Częstość skumulowana", color='red')
        ax2.tick_params(axis='y', colors='red')
        ax2.set_ylim(0, 1)
        ax.set_title(title)
        ax.grid(axis='y', linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()




# Wywołanie poprawionego wykresu
plot_vei_counts([vei_all, vei_confirmed, vei_uncertain])
```

```python
# YOUR CODE HERE
evidence_data = (eruptions_data.groupby(["Evidence Type", "Evidence Method", "VEI"]).size().unstack(level=2, fill_value=0)) #unstack ustawia VEI jako kolumne, a pozostałe dwa to dwupoziomowe nagłówki
evidence_data.columns = evidence_data.columns.astype(int) #VEI po unstack jest traktowane jako float
evidence_data = evidence_data.sort_index(axis=1, ascending=True) #posrtowanie kolumny VEI
evidence_data["All"] = evidence_data.sum(axis=1) #dodanie kolumny, suma wszystkich erupcji
evidence_data = evidence_data.astype(int) #bez miejsc po przecinku, musi byc wyzej fill_value=0 zeby moglo zadzialac
evidence_data
```
